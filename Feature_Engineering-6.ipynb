```python
Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.



ANS-1


To calculate the Pearson correlation coefficient (also known as Pearson's r) between two variables, you need to follow these steps:

Step 1: Organize the data into two separate lists or arrays: one for the amount of time students spend studying for the exam and another for their corresponding final exam scores.

Step 2: Compute the mean (average) of each list.

Step 3: Calculate the covariance between the two variables using the formula:

Cov(X, Y) = Σ[(Xᵢ - X̄)(Yᵢ - Ȳ)] / (n - 1)

where:
- X and Y are the two variables (time spent studying and exam scores).
- Xᵢ and Yᵢ are individual data points for time spent studying and exam scores, respectively.
- X̄ and Ȳ are the means of time spent studying and exam scores, respectively.
- n is the number of data points in the dataset.

Step 4: Calculate the standard deviation of each variable:

σ(X) = √[Σ(Xᵢ - X̄)² / (n - 1)]
σ(Y) = √[Σ(Yᵢ - Ȳ)² / (n - 1)]

Step 5: Finally, compute the Pearson correlation coefficient (r):

r = Cov(X, Y) / (σ(X) * σ(Y))

Interpretation of the result:
The Pearson correlation coefficient, r, ranges between -1 and 1. Here's what different values of r indicate about the relationship between the two variables:

- r = 1: A perfect positive correlation. This means that when one variable increases, the other also increases in a linear fashion. In this context, it would mean that as students spend more time studying for the exam, their final exam scores tend to increase linearly.
- r = -1: A perfect negative correlation. This indicates that when one variable increases, the other decreases in a linear fashion. In this context, it would mean that as students spend more time studying for the exam, their final exam scores tend to decrease linearly.
- r ≈ 0: Little to no linear correlation. This suggests that there is no strong linear relationship between the time spent studying and the final exam scores. However, it's important to note that there might still be other types of relationships (non-linear) between the variables that the Pearson correlation coefficient cannot capture.

Remember that the Pearson correlation coefficient only measures linear relationships and does not indicate causation. It gives an idea of how two variables move together in a linear manner, but it does not imply that one variable causes the other to change. Always consider the context of the data and the nature of the relationship when interpreting the correlation coefficient.
                                                  
                                                  
                                                  
Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables.
Suppose you have collected data on the amount of sleep individuals get each night and their overall job
satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
variables and interpret the result.
                                                  
  
                                                  
                                                  
ANS-2
                                                  
                                                  
                                                  
 To calculate Spearman's rank correlation coefficient (often denoted by ρ or rs) between two variables, such as the amount of sleep individuals get each night and their overall job satisfaction level (measured on a scale of 1 to 10), you need to follow these steps:

Step 1: Organize the data into two separate lists or arrays: one for the amount of sleep and another for the corresponding job satisfaction levels.

Step 2: Assign ranks to each data point in both lists separately. Rank 1 to the smallest value, rank 2 to the next smallest, and so on. If there are ties (i.e., multiple individuals with the same sleep duration or job satisfaction level), assign the average rank to all tied values.

Step 3: Calculate the differences between the ranks for each individual in both lists.

Step 4: Square each of the rank differences.

Step 5: Compute the Spearman's rank correlation coefficient using the formula:

ρ = 1 - [6 * Σ(d²) / (n * (n² - 1))]

where:
- d is the difference between the ranks for each individual.
- Σ denotes the sum of the values.
- n is the number of data points in the dataset.

Interpretation of the result:
The Spearman's rank correlation coefficient, ρ, ranges between -1 and 1:

- ρ = 1: A perfect monotonic positive correlation. This indicates that as the amount of sleep individuals get each night increases, their overall job satisfaction level also increases, following a monotonic (but not necessarily linear) pattern. In other words, individuals who sleep more tend to have higher job satisfaction levels, and vice versa.
- ρ = -1: A perfect monotonic negative correlation. This suggests that as the amount of sleep individuals get each night increases, their overall job satisfaction level decreases, following a monotonic pattern. In this case, individuals who sleep less tend to have higher job satisfaction levels, and vice versa.
- ρ ≈ 0: Little to no monotonic correlation. This means that there is no strong monotonic relationship between the amount of sleep and job satisfaction levels. However, it's essential to remember that this does not rule out the possibility of other types of relationships (such as non-monotonic or nonlinear) between the variables.

Spearman's rank correlation assesses the monotonic relationship between two variables, meaning that it measures whether the variables tend to move together in a consistent direction (either increasing or decreasing) without necessarily being linearly related. Like Pearson correlation, Spearman's rank correlation also does not imply causation, but it gives insights into the direction and strength of the monotonic association between the two variables.
  
                                                  
                                                  
                                                  
Q3. Suppose you are conducting a study to examine the relationship between the number of hours of
exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables
for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation
between these two variables and compare the results.
                                                  
                                                  
                                                  
                                                  
 ANS-3
                                                  
                                                  
                                                  
  To calculate both the Pearson correlation coefficient and the Spearman's rank correlation between the number of hours of exercise per week and body mass index (BMI) in a sample of adults, you need to follow these steps:

Step 1: Organize the data into two separate lists or arrays: one for the number of hours of exercise per week and another for the corresponding BMI values.

Step 2: Compute the means (average) of both lists.

Step 3: Calculate the covariance between the two variables using the Pearson correlation formula:

Cov(X, Y) = Σ[(Xᵢ - X̄)(Yᵢ - Ȳ)] / (n - 1)

where:
- X and Y are the two variables (number of hours of exercise and BMI).
- Xᵢ and Yᵢ are individual data points for the number of hours of exercise and BMI, respectively.
- X̄ and Ȳ are the means of the number of hours of exercise and BMI, respectively.
- n is the number of data points in the dataset (in this case, n = 50).

Step 4: Calculate the standard deviations of both variables:

σ(X) = √[Σ(Xᵢ - X̄)² / (n - 1)]
σ(Y) = √[Σ(Yᵢ - Ȳ)² / (n - 1)]

Step 5: Compute the Pearson correlation coefficient (r) using the formula:

r = Cov(X, Y) / (σ(X) * σ(Y))

Step 6: Calculate the ranks for both variables separately.

Step 7: Compute the differences between the ranks for each participant in both lists.

Step 8: Square each of the rank differences.

Step 9: Calculate the Spearman's rank correlation coefficient (ρ) using the formula:

ρ = 1 - [6 * Σ(d²) / (n * (n² - 1))]

where:
- d is the difference between the ranks for each participant.
- Σ denotes the sum of the values.
- n is the number of data points in the dataset (in this case, n = 50).

Now, compare the results:

The Pearson correlation coefficient (r) measures the linear relationship between the two variables. If r is close to 1, it indicates a strong positive linear correlation, meaning that as the number of hours of exercise per week increases, BMI tends to decrease (or vice versa). If r is close to -1, it indicates a strong negative linear correlation, meaning that as the number of hours of exercise per week increases, BMI tends to increase (or vice versa). If r is close to 0, it suggests little to no linear correlation between the two variables.

On the other hand, the Spearman's rank correlation coefficient (ρ) measures the monotonic relationship between the two variables. If ρ is close to 1, it indicates a strong positive monotonic correlation, suggesting that as the number of hours of exercise per week increases, BMI tends to decrease (or vice versa). If ρ is close to -1, it indicates a strong negative monotonic correlation, suggesting that as the number of hours of exercise per week increases, BMI tends to increase (or vice versa). If ρ is close to 0, it suggests little to no monotonic correlation between the two variables.

Comparing the results of both coefficients will give you insights into the type and strength of the relationship between the number of hours of exercise per week and BMI in the sample of adults. Keep in mind that Pearson correlation is more sensitive to linear relationships, while Spearman's rank correlation is more robust to non-linear relationships and outliers.
                                                  
                                                  
                                                  
                                                  
                                                  
                                                  
Q4. A researcher is interested in examining the relationship between the number of hours individuals
spend watching television per day and their level of physical activity. The researcher collected data on
both variables from a sample of 50 participants. Calculate the Pearson correlation coefficient between
these two variables.
                                                  
                                                  
                                                  
 ANS-4
                                                  
                                                  
 To calculate the Pearson correlation coefficient (also denoted as r) between the number of hours individuals spend watching television per day and their level of physical activity, you need to follow these steps:

Step 1: Organize the data into two separate lists or arrays: one for the number of hours spent watching television per day and another for the corresponding level of physical activity.

Step 2: Compute the means (average) of both lists.

Step 3: Calculate the covariance between the two variables using the Pearson correlation formula:

Cov(X, Y) = Σ[(Xᵢ - X̄)(Yᵢ - Ȳ)] / (n - 1)

where:
- X and Y are the two variables (number of hours watching TV and level of physical activity).
- Xᵢ and Yᵢ are individual data points for the number of hours watching TV and physical activity, respectively.
- X̄ and Ȳ are the means of the number of hours watching TV and physical activity, respectively.
- n is the number of data points in the dataset (in this case, n = 50).

Step 4: Calculate the standard deviations of both variables:

σ(X) = √[Σ(Xᵢ - X̄)² / (n - 1)]
σ(Y) = √[Σ(Yᵢ - Ȳ)² / (n - 1)]

Step 5: Compute the Pearson correlation coefficient (r) using the formula:

r = Cov(X, Y) / (σ(X) * σ(Y))

Once you have calculated the values of Cov(X, Y), σ(X), and σ(Y), you can plug them into the formula to find the Pearson correlation coefficient (r).

Interpretation of the result:
The Pearson correlation coefficient, r, ranges between -1 and 1:

- r = 1: A perfect positive correlation. This means that there is a strong positive linear relationship between the number of hours individuals spend watching television per day and their level of physical activity. As the number of hours spent watching TV increases, physical activity also increases in a linear manner.
- r = -1: A perfect negative correlation. This indicates a strong negative linear relationship, suggesting that as the number of hours spent watching TV increases, the level of physical activity decreases in a linear manner.
- r ≈ 0: Little to no linear correlation. This suggests that there is little or no linear relationship between the number of hours spent watching TV and the level of physical activity. In other words, watching TV and physical activity are not strongly related in a linear manner.

Keep in mind that the Pearson correlation coefficient only measures linear relationships and may not capture more complex or non-linear relationships between the variables. Additionally, correlation does not imply causation, so even if there is a correlation between the two variables, it does not mean that one causes the other to change. It's essential to interpret the results within the context of the study and consider other factors that may influence the relationship between TV watching and physical activity.
                                                  
                                                  
                                                  
                                                  
Q5. A survey was conducted to examine the relationship between age and preference for a particular
brand of soft drink. The survey results are shown below:
                                                  
                                                  
                                                  
 ANS-5
                                                  
                                                  
    To examine the relationship between age and preference for a particular brand of soft drink, we can use a frequency table to organize the data and calculate the mode for the preference. Since the ages for two individuals are missing, we cannot perform a comprehensive statistical analysis. However, we can summarize the data and provide some insights based on the available information:

Based on the survey results:

Soft drink preference:
- Coke: 3 participants
- Pepsi: 2 participants
- Mountain Dew: 1 participant

Age (Years):
- 25
- 42
- 37
- 19 (missing preference)
- 31
- 28 (missing preference)

Based on the data, it appears that Coke is the most preferred brand of soft drink among the surveyed participants, with 3 individuals selecting it. Pepsi is the second most preferred brand, chosen by 2 participants, while Mountain Dew is preferred by 1 participant.

Regarding age, we have information for 4 individuals, with ages of 25, 42, 37, and 31. The ages of 19 and 28 are available but do not have corresponding soft drink preferences listed.

Since we have limited data, we cannot calculate correlation coefficients or perform a more in-depth statistical analysis to determine any significant relationships between age and soft drink preference. Additional data points would be required to draw any meaningful conclusions or make generalizations.

To further investigate the relationship between age and soft drink preference, a more extensive and representative survey with a larger sample size would be necessary. This would allow for a better understanding of potential patterns or trends that may exist between age groups and their preferred soft drinks.
                                                  
                                                  
                                                  
                                                  
                                                  
                                                  
```
